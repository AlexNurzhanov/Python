`f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30`


1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c, e, k = -12, -18, 5, 10, -30
x = np.arange(-50, -20, 0.001)
x_cut = np.arange(-45, -42, 0.001)
```

```python
def func(x):
    function = a*np.sin(np.cos(x))*x**4 + b*x**3 + c*x**2 + e*x + k
    return function

root = -45
for i in np.arange(-45, -42, 0.001):
    if np.absolute(func(root))>np.absolute(func(i)):
        root = i

min_func = min(func(x_cut))
def extr_func(x, min_func):
    extremum_function = a*np.sin(np.cos(x))*x**4 + b*x**3 + c*x**2 + e*x + k - min_func
    return  extremum_function

extr_x_cut = -45
for i in np.arange(-45, -42, 0.001):
    if np.absolute(extr_func(extr_x_cut, min_func))>np.absolute(extr_func(i, min_func)):
        extr_x_cut = i

x_range_down = np.arange(-45, extr_x_cut, 0.001)
x_range_up = np.arange(extr_x_cut, -42, 0.001)
```

```python
plt.title(f'Корень функции на участке (-45; -42): {round(root, 3)}')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.minorticks_on()
plt.grid(which='major',
        color = 'k', 
        linewidth = 0.5)
plt.grid(which='minor', 
        color = 'k', 
        linestyle = ':')
plt.plot(x, func(x), 'b')
plt.plot(x_range_down, func(x_range_down), 'r', label="Убывание")
plt.plot(x_range_up, func(x_range_up), 'g', label="Возрастание")
plt.text(extr_x_cut, func(extr_x_cut) + 30, f'Вершина функции x = {round(extr_x_cut, 3)}')
plt.legend()
plt.show()
```
